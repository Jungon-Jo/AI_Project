ReLU 함수

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
# 그래프 그리기
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.5, 5.5)  # y축 범위 설정
plt.title("ReLU Activation Function")
plt.grid()
plt.show()
```
