```python
# ===================================================================

# 문제 1

text = "Machine Learning is the FUTURE of AI!"

# 문자열에서 대문자는 소문자로, 소문자는 대문자로 변환하세요.

# 변환 후, 결과 문자열에서 !를 제거하세요.

# -------------------------------------------------------------------

# 문제 2

sentence = "Data analysis and data science are important fields."

# 문자열에서 data라는 단어의 모든 시작 위치(인덱스)를 출력하세요.

# data가 몇 번 등장하는지 대소문자 구분 없이 계산하세요.

# -------------------------------------------------------------------

# 문제 3

text = "    ***Welcome to Python!***    "

# 문자열에서 앞뒤의 공백과 *를 제거하세요.

# 결과 문자열에서 공백을 _로 치환하세요.

# -------------------------------------------------------------------

# 문제 4

sentence = "Learn Python, Java, C++ and JavaScript"

# split을 사용해 언어 이름을 추출하세요.

# 모든 언어를 역순으로 연결해 "->"로 구분된 문자열을 만드세요.

# -------------------------------------------------------------------

# 문제 5

data = {"name": "Alice", "score": 95.6789}

# 이름과 점수를 다음 형식으로 출력하세요:
# Name  : Alice
# Score : 95.68

# 점수는 소수점 둘째 자리까지만 표시하고, : 뒤로는 공백을 일정하게 맞추세요.

# -------------------------------------------------------------------

# 문제 6

text = "Python\tis\nfun\tand\npowerful"

# 문자열을 탭(\t)과 줄바꿈(\n)을 기준으로 분리해 리스트로 변환하세요.

# 분리된 리스트를 다시 줄바꿈과 탭을 조합하여 하나의 문자열로 만드세요.

# -------------------------------------------------------------------

# 문제 7

numbers = [10, 15, 20, 25, 30]

# 리스트에서 15 이상 25 이하인 숫자만 추출하여 새로운 리스트를 생성하세요.
# 새 리스트에 숫자가 모두 짝수인지 확인하세요.

# -------------------------------------------------------------------

# 문제 8

numbers = [4, 2, 9, 1, 5, 6]

# 숫자들을 오름차순으로 정렬한 뒤, 가장 작은 두 개의 숫자를 제거하세요.
# 남은 숫자들을 거꾸로 뒤집어 출력하세요.

# -------------------------------------------------------------------

# 문제 9

items = [3, 6, 3, 7, 9, 6, 1, 3]

# 중복된 값을 제거하고, 오름차순으로 정렬한 리스트를 만드세요.

# -------------------------------------------------------------------

# 문제 10

list1 = [1, 2, 3]
list2 = [4, 5, 6]

# 두 리스트를 합쳐 새로운 리스트를 생성하되, 각 요소에 +10을 더한 결과를 포함하세요.

# -------------------------------------------------------------------

# 문제 11

scores = [88, 92, 76, 81, 95]

# 점수가 90 이상인 경우 "A"로, 80 이상 90 미만이면 "B"로, 나머지는 "C"로 변환한 새로운 리스트를 생성하세요.

# -------------------------------------------------------------------

# 문제 12

numbers = [10, 15, 20, 25, 30, 35]

# 리스트에서 20과 30을 제거한 결과를 출력하세요.

# ===================================================================

# 풀이

# 문제 1

transformed = "".join([char.lower() if char.isupper() else char.upper() for char in text])
result = transformed.replace("!", "")
print(result)  # "mACHINE lEARNING IS THE future OF ai"

# -------------------------------------------------------------------

# 문제 2

indices = [i for i in range(len(sentence)) if sentence.lower().startswith("data", i)]
count = sentence.lower().count("data")
print(indices)  # [0, 18]
print(count)  # 2

# -------------------------------------------------------------------

# 문제 3

cleaned = text.strip(" *").replace(" ", "_")
print(cleaned)  # "Welcome_to_Python!"

# -------------------------------------------------------------------

# 문제 4

languages = sentence.split(" ")[1::2]
result = "->".join(languages[::-1]) 
print(result)  # "JavaScript->C++->Java->Python"

# -------------------------------------------------------------------

# 문제 5

formatted = f"Name  : {data['name']:<10}\nScore : {data['score']:.2f}"
print(formatted)

# -------------------------------------------------------------------

# 문제 6

parts = text.replace("\t", "\n").split("\n")
recombined = "\t".join(parts)
print(recombined)  # "Python\tis\tfun\tand\tpowerful"

# -------------------------------------------------------------------

# 문제 7

filtered = [num for num in numbers if 15 <= num <= 25]
all_even = all(num % 2 == 0 for num in filtered)
print(filtered)  # [15, 20, 25]
print(all_even)  # False

# -------------------------------------------------------------------

# 문제 8

numbers.sort()
numbers = numbers[2:]
numbers.reverse()
print(numbers)  # [9, 6, 5]

# -------------------------------------------------------------------

# 문제 9

unique_items = sorted(set(items))
print(unique_items)  # [1, 3, 6, 7, 9]

# -------------------------------------------------------------------

# 문제 10

combined = [x + 10 for x in list1 + list2]
print(combined)  # [11, 12, 13, 14, 15, 16]

# -------------------------------------------------------------------

# 문제 11

grades = ["A" if score >= 90 else "B" if score >= 80 else "C" for score in scores]
print(grades)  # ["B", "A", "C", "B", "A"]

# -------------------------------------------------------------------

# 문제 12

filtered = [num for num in numbers if num not in (20, 30)]
print(filtered)  # [10, 15, 25, 35]

# ===================================================================
```
