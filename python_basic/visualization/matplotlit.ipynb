Matplotlib

```python
# plt.plot()
# 선 그래프 생성
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../dataset/diabetes.csv", encoding="CP949")
data = df.loc[:, "Pregnancies"].values.tolist()
min_value = min(data) # 데이터의 최솟값
max_value = max(data) # 데이터의 최댓값
normalized = [] # 정규화된 값을 저장 할 리스트
for num in data:
  # 정규화 공식 적용
  norm_value = (num - min_value) / (max_value - min_value)
  normalized.append(norm_value)
data_index = []
for num in range(0, 768):
  data_index.append(num)
  
plt.plot(data_index, normalized)
plt.show()
```

```python
# plt.scatter()
# 산점도 생성
plt.scatter(data_index, normalized)
plt.show()
```

```python
# plt.bar()
# 막대그래프 생성
plt.bar(data_index, height=normalized)
plt.show()
```

```python
# plt.hist()
# 히스토그램 생성
plt.hist(normalized, bins = 50)
plt.show()
```

```python
# plt.boxplot()
# 상자 그림 생성
plt.boxplot(normalized)
plt.show()
```

```python
# plt.pie()
# 원 그래프 생성
plt.pie(normalized)
plt.show()
```

```python
# plt.subplot()
# 여러개의 그래프를 하나의 Figure에 배치
plt.subplot(2, 2, 1)
plt.show()
```

```python
# plt.xlabel()
# plt.ylabel()
# x축 레이블 설정
# y축 레이블 설정

plt.xlabel("Index")
plt.ylabel("Value")
```

```python
# plt.title()
# 그래프 제목 설정
plt.title("Title")
```
