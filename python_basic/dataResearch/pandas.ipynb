Pandas 함수 예시

```python
# read_csv()
# CSV 파일 읽기
import pandas as pd
df = pd.read_csv("../dataset/diabetes.csv", encoding="CP949")
print(df)
```

```python
# head(), tail()
# 데이터프레임의 처음 또는 마지막 부분 확인
df.head()
df.tail()
```

```python
# shape
# 데이터프레임의 행과 열 개수 확인
df.shape # 출력 : (3행 4열)
```

```python
# info()
# 데이터프레임의 요약 정보 확인
df.info()
```

```python
# describe()
# 데이터프레임의 기본적인 통계 정보 확인
df.describe()
```

```python
# loc[]
# 데이터 선택
df.loc[:, "Pregnancies"]
df["Pregnancies"]
# loc[:, "컬럼명"]
# df["컬럼명"]
```

```python
# iloc[]
# 데이터 선택
df.iloc[1] # 하나의 행값 전체를 출력
```

```python
# groupby
# 데이터 그룹화
df.groupby("Pregnancies").mean()
```

```python
# sort_values()
# 데이터 정렬
df.sort_values("Pregnancies")
```

```python
# fillna()
# 결측치 채우기
df.fillna(0)
```

```python
# dropna()
# 결측치 행 삭제
df.dropna()
```
